```python
df1['birth_date'] = df1['birth_date'].replace('NaN', pd.NA)  # NaN değerleri gerçek NaN olarak ayarla
df1['birth_date'] = pd.to_datetime(df1['birth_date'], format='%Y', errors='coerce')  # Tarih sütununu datetime formatına çevir

# Doğum yıllarını elde etme ve gruplama
df1['birth_year'] = df1['birth_date'].dt.year
birth_year_bins = list(range(1900, 2021, 20))  # Her 20 yılda bir grupla
df1['birth_year_group'] = pd.cut(df1['birth_year'], bins=birth_year_bins, right=False)

# Her grup için doğum yılının ortalamasını hesaplama
group_means = df1.groupby('birth_year_group')['birth_year'].mean()

# NaN değerleri grup ortalamaları ile doldurma fonksiyonu
def fill_birth_year(row):
    if pd.isna(row['birth_year']):
        group = row['birth_year_group']
        if pd.notna(group) and group in group_means:
            return group_means[group]
    return row['birth_year']

df1['birth_year'] = df1.apply(fill_birth_year, axis=1)

# Doldurulmuş doğum yıllarını tekrar datetime formatına çevirme
df1['birth_date'] = pd.to_datetime(df1['birth_year'], format='%Y')

# Sonuçları kontrol etme
print(df1[['birth_date', 'birth_year', 'birth_year_group']])
```

```python
def fill_birth_dates(df1):
    df1['birth_date'] = pd.to_datetime(df1['birth_date'])
    avg_birth_dates = df1.groupby('category')['birth_date'].apply(lambda x: x.mean()).to_dict()
    df1['birth_date'] = df1.apply(lambda row: avg_birth_dates[row['category']] if pd.isnull(row['birth_date']) else row['birth_date'], axis=1)
    return df1

# 'birth_year', 'birth_year_group' ve 'interval' sütunlarını çıkararak boş doğum tarihlerini doldur
df1 = df1.drop(['birth_year', 'birth_year_group'], axis=1)
df1_filled = fill_birth_dates(df1)
df1_filled
```

```python
def fill_death_dates(df1):
    df1['death_date'] = pd.to_datetime(df1['death_date'])
    avg_death_dates = df1.groupby('category')['death_date'].apply(lambda x: x.mean()).to_dict()
    df1['death_date'] = df1.apply(lambda row: avg_death_dates[row['category']] if pd.isnull(row['death_date']) else row['death_date'], axis=1)
    return df1

# 'death_date' sütunundaki boş değerleri doldur
df1_filled = fill_death_dates(df1)
df1_filled
```

### <span style="color:blue"> 5) Nobel Ödülü kazananların cinsiyetlerini, yaşlarını, ödül kategorisi ve yılları kullanarak görselleştiriniz.(Her bir ödül kategorisi için ayrı grafik gösteriniz) Çıkan sonuçlara göre görseli yorumlayınız.

```python
# Tarih sütunlarını datetime formatına çevirme
df1['birth_date'] = pd.to_datetime(df1['birth_date'], errors='coerce')
df1['death_date'] = pd.to_datetime(df1['death_date'], errors='coerce')

# Yaş hesaplama
df1['age'] = (df1['death_date'] - df1['birth_date']).dt.days // 365.25

# Geçerli yaş değerlerine sahip verileri filtreleme (yaş > 0)
df1_age_filtered = df1[(df1['age'] > 0)]

# Görselleştirme
plt.figure(figsize=(14, 8))
sns.violinplot(x='category', y='age', hue='sex', data=df1_age_filtered, split=True, inner='quart')
plt.title('Nobel Ödülü Kazananların Cinsiyet ve Ödül Kategorisine Göre Yaş Dağılımı')
plt.xlabel('Kategori')
plt.ylabel('Yaş')
plt.legend(title='Cinsiyet')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### <span style="color:blue"> 6) 1938-1945 yılı arasında Nobel Ödülü kazananların kategorilerini ve ülkelerini görselleştirip yorumlayınız.
